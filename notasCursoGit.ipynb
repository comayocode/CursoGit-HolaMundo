## ¿QUÉ ES?
Es una herramienta que nos permite tener almacenado todo el historial del código que hemos desarrollado. Sirve bastante a la hora de tener algún bug o error en el código, cada que realizamos un cambio a nuestro código subimos esos cambios a un repositorio en la nube, por poner un ejemeplo... el primer cambio sería la versión 1.0, al realizar un cambio subimos dichos cambios a un repositorio como la si fuera la versión 1.1; si por algún motivo en la versión 1.1 hay algún error en el código, simplemente hacemos un ROLLBACK que consiste en trabajar sobre la versión 1.0.

## CONFIGURACIÓN DE GIT

### Configuración global
Sirve para que la configuración acute de forma global y no por proyecto

- Establecer el nombre de usuario, si se va a poner espacios en el nombre de tiene que usar ""

    `git config --global user.name`


- Establecer el correo electrónico

    `git config --global user.email`

- Establecer el editor de código

    `git config --global core.editor "code --wait"`

- core.autocrlf

    En windows al hacer un salto de linea se agregan 2 caracteres especiales "Carriage Return (CR)" & "Line Feed (LF)". En linux/MAC solo se agrega LF. El desarrollador de windows al subir código necesita eliminar el caracter CR, y si quiere descargar código del repositorio debe agregar el caracter CR, para ello se requiere cambiar la configuración del crlf, esta configuración debe ser TRUE, mientras que en MAC debe ser INPUT

    WINDOWS : `git config --global core.autocrlf true`

    MAC/LINUX : `git config --global core.autocrlf input`

**Si se quiere indagar más a fondo sobre las configuraciones de git se escribe:**  `git config -h`

## FLUJO DE TRABAJO

1. Computador

    Cuando agregamos y modificamos un archivo de código en el computador no quiere decir que estos se agregen automáticamente a un repositorio, para ello se tiene que ejecutar el comando `git add`. Cuando se ejecuta el comando se seleccionan los archivos que se quieren agregar a una etapa que se llama **Stege**
2. Stege

    Al pasar los archivos a esta etapa se verifican todos los cambios qeu se quieran pasar al repositorio. Esto no quiere decir que al estar en la etapa de **Stege** se agregan los archivos al repositorio, es como una etapa intermedia en la que indicamos cuales son los cambios que se pasen al repositorio (no pasan todos los cambios, sólo los que seleccionemos).
3. Commit

    Ya teniendo los cambios que se quieren subir al repositorio se ejecuta el comando `git commit` dejando un mensaje o descripción con los cambios realizados.
4. Server

    Por ultimo se usa el comando `git push` para subir los cambios a un servidor, puede ser Github, Gitlab, entre otros...

>En caso de eliminar un archivo debemos ejecutar el comando `git add` (se agrega un archivo que hemos eliminado). Esto sirve para mantener el registro de cada cambio que realizamos en el proyecto

>Otro comando: `code .` se usa para abrir una nueva ventana del editor con la carpeta. Previamente se tiene que estar dentro de la ruta/carpeta en la terminal

`Untracked files` : Archivos que git no les está haciendo seguimiento

Por defecto git no sigue/hace seguimiento a todos los archivos, para ello se tiene que usar el comando `git add (varios parametros)`:

- `git add (nombre archivo)` **Ejemplo** : `git add archivo.txt` : Añadir un archivo en específico.

- `git add (nombre archivo) (nombre archivo)` **Ejemplo** : `git add archivo1.txt archivo2.txt` : Agregar varios archivos

- `git add (extensión del archivo)` **Ejemplo** : `git add .txt`: Añade todos los archivos que terminan con esa extensión.

- `git add .` : Añade todos los archivos (Esto es una mala práctica y no es recomendable ya que podemos añadir accidentalmente archivos que no quisieramos añadir).

## Ramas en git o brenches

Hasta ahora hemos visto la forma lineal de trabajar con GIT, es decir, cada vez que realizamos un cambio este se verá reflejado por encima de nuestro historial.

<img src="img/Screenshot_46.png">

Ahora, si hay más de 1 desarrollador trabajando en un mismo proyecto se requiere crear una rama o "brench" secundaria para poder trabajar en el código que estemos desarrollando en nuestra máquina, para ello "sacamos" o "creamos" una rama secundario a la cual se le estará agregando commits (esta es completamente independiente de la rama principal), y finalmente terminada la función/código volverla a unir con la rama principal.

<img src="img/Screenshot_47.png">

#### Crear la rama

Antes de crear una rama tenemos que **VERIFICAR** en que rama nos encontramos usando. `git branch` nos indicará que estamos en la rama "MASTER", que es la misma rama "MAIN", sólo cambia la semántica.

<img src="img/Screenshot_52.png">

Usamos `git checkout -b (Nombre de la rama)` **(Hay varias nomenclaturas, eso depende de la empresa)**, en nuestro caso pondremos como nombre "ramab" `git checkout -b ramab`.

<img src="img/Screenshot_53.png">

Ejecutamos `git branch` para verificar que estamos en la **ramab** que creamos.

<img src="img/Screenshot_54.png">

Ahora vamos a hacer cambios sobre el archivo. Posteriormente añadimos el archivo a **Stage** y hacemos el **commit**:

<img src="img/Screenshot_55.png">

Vemos el historial con `git log --oneline`. Nos indica que estamos en la cabeza de la **ramab**

<img src="img/Screenshot_56.png">

Usamos `cat archivo2.txt` para ver el contenido del archivo

<img src="img/Screenshot_57.png">

Ahora nos moveremos a la rama principal `git checkout master`(main, master) y de nuevo ejecutamos el comando `cat archivo2.txt`. Podemos observar que ahora estamos trabajando sobre la rama principal, pero en esta aún no se ven reflejados los cambios de la **ramab** (línea 4 de la siguiente imagen).

<img src="img/Screenshot_58.png">

#### Unir la ramab con la rama principal

Primero debemos verificar que estamos trabajando en la rama main/master. Después ejecutamos el comando `git merge (nombre de la rama)`, con el nombre de la rama que querenmos traer a nuestra rama principal, en nuestro caso sería `ramab`.

<img src="img/Screenshot_59.png">

Ahora verificamos que si se trajo/unió la `ramab` con la rama principal o `master`, ejecutamos `cat archivo2.txt`

<img src="img/Screenshot_60.png">

Esto es útil para que en un equipo cada desarrollador cree su propia rama, implemente su funcionalidad y luego la mezcla haciendo un `merge` con la rama principal.

### **RESUMEN**

>Se verifica que la zona de trabajo esté limpia

>Crear una rama usando `git checkout -b (Nombre de la rama)`

>Se hacen los respectivos cambios o nuevas implementaciones

>Para unir la ramab con la master primero tenemos que estar en la rama master `git checkout master` y finalmente la unimos con `git merge ramab`
